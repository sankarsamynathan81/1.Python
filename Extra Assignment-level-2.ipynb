```python
#Print Odd numbers in the list 
a = [20,10,16,19,25,1,276,188]

for i in a:
    if i % 2 != 0:
        print(i,"is Odd")
```

```python
#Print Even  numbers in the list 
Even = [20,10,16,19,25,1,276,188]

for i in Even:
    if i % 2 == 0:
        print(i,"is Even")
```
