```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
ageVal=(int)(input("Enter your Age"))
print("age:",ageVal)
if ageVal < 18: 
    print ("Children")
elif  ageVal >=18 and ageVal <= 23:
    print ("Adult")
elif  ageVal >=23 and ageVal <= 60:
    print ("Citizen")
elif  ageVal >=60 and ageVal <= 80:
    print ("senior citizen")
```

```python
#Check whether number is Divible by 5 
divValue=(int)(input("Value="))
# Extract the last digit using N % 10
last_digit = divValue % 10
# Check for the last digit
if last_digit == 0 or last_digit == 5:
    print(divValue," is divisible by 5")
else:
    print(divValue," is not divisible by 5")
```
